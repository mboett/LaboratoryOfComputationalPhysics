1\. **Radioactive decay chain**

${\rm Tl}^{208}$ decays to ${\rm Pb}^{208}$ with a half-lieve of 3.052 minutes. Suppose to start with a sample of 1000 Thallium atoms and 0 of Lead atoms.

* Take steps in time of 1 second and at each time-step decide whether each Tl atom has decayed or not, accordingly to the probability $p(t)=1-2^{-t/\tau}$. Subtract the total number of Tl atoms that decayed at each step from the Tl sample and add them to the Lead one. Plot the evolution of the two sets as a function of time  
* Repeat the exercise by means of the inverse transform method: draw 1000 random numbers from the non-uniform probability distribution $p(t)=2^{-t/\tau}\frac{\ln 2}{\tau}$ to represent the times of decay of the 1000 Tl atoms. Make a plot showing the number of atoms that have not decayed as a function of time


```python
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import gamma

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%precision 4

Tl = 1000
Pb = 0
tau = 3.052*60
Tl_Ev = np.array([Tl])
Pb_Ev = np.array([Pb])


# First point
def p(t):
    return (1-2**(-t/tau))

prob = p(1)
while Tl>0:  
    for i in range(Tl):
        ran = np.random.uniform(0,1)
        if ran <= prob:
            Tl=Tl-1
            Pb=Pb+1
    Tl_Ev = np.append(Tl_Ev, Tl)
    Pb_Ev = np.append(Pb_Ev, Pb)

fig = plt.figure(figsize=(12,12)) 
ax1 = fig.add_subplot(211)   
ax1.plot(Tl_Ev, label = 'Tl atom')
ax1.plot(Pb_Ev, label = 'Pb atom')

ax1.set_xlabel('Time')
ax1.set_ylabel('# of atom')
ax1.set_title('Method 1')
ax1.legend()

# Second point
def expon_icdf(p, mu=np.log(2)/tau):
    """Inverse CDF of exponential distribution - i.e. quantile function."""
    return -np.log(1-p)/mu

Tl = 1000
u = np.random.random(Tl)
v = expon_icdf(u)

ax2 = fig.add_subplot(212)
ax2.hist(v, bins=100, cumulative=-1, label='Tl atom')

ax2.set_title('Method 2')
ax2.set_xlabel('Time')
ax2.set_ylabel('# of atom')
ax2.legend()

plt.show()
```

2\. **Rutherford Scattering**

The scattering angle $\theta$ of $\alpha$ particles hitting a positively charged nucleus of a Gold atom ($Z=79$) follows the rule:

$$
\tan{\frac{1}{2} \theta} = \frac{Z e^2} {2\pi \epsilon_0 E b}
$$

where $E=7.7$ MeV and $b$ beam is the impact parameter. The beam is represented by a 2D gaussian distribution with $\sigma=a_0/100$ for both coordinates ($a_0$ being the Bohr radius). Assume 1 million $\alpha$ particles are shot on the gold atom.

Computing the fraction of particles that "bounce back",i.e. those particle whose scattering angle is greater than $\pi/2$ (which set a condition on the impact parameter $b$)

elementary charge = 1.16 * 10^(-19)

epsilon zero = 8.85 * 10^(-12)

a0 = 5.29 * 10^(-11)

```python
# Constants
num_shot =10**6
Z = 79
ec = 1.16*(10**-19)
E = 7.7*(10**6)*ec
eps0 = 8.85*(10**-12)
a0 = 5.29*(10**-11)

# Box-Muller method
z = np.random.rand(num_shot)
b = np.sqrt(-2*(a0/100)**(2)*np.log(1-z))
theta = 2*np.arctan(Z*ec*ec/(2*np.pi*eps0*E*b))

frac = theta[theta>np.pi/2].shape[0]/num_shot
print('Fraction of particicles bounced back:', frac)
```

3\. **Monte Carlo integration: hit/miss vs mean value method**

Consider the function 

$$f(x) =\sin^2{\frac{1}{x(2-x)}}$$

* Compute the integral of $f(x)$ between 0 and 2 with the hit/miss method. Evaluate the error of your estimate
* Repeat the integral with the mean value method. Evaluate the error and compare it with the previous one

```python
def f(x):
    return np.sin(1/(2*x-x**2))**2

true_res = 1.4514
print('True result:', true_res )

# Hit/Miss method
N = 100000
count = 0
for i in range(N):
    x = 2*np.random.random()
    y = np.random.random()
    if y<f(x): 
        count+=1
I_HM = 2*count/N
print('Hit/Miss method result:', I_HM, ' Error:', abs(I_HM-true_res))

# Mean method
f_ran = f(2*np.random.rand(N))
I_M = 2*f_ran.sum()/N
print('Mean method result:', I_M, ' Error:', abs(I_M-true_res))
```

4\. **Monte Carlo integration in high dimension**

* Start of by computing the area of a circle of unit radius, by integrating the function 

$$
f(x,y)=
\left\{
\begin{array}{ll}
      1 & x^2+y^2\le 1 \\
      0 & {\rm elsewhere}
\end{array} 
\right.
$$

* Generalize the result for a 10D sphere



```python
# Unit radius circle
def circle(xy):
    if xy[0]**2+xy[1]**2<1:
        return 1
    else:
        return 0
    
# Mean Montecarlo integration
N = 100000
xy_ran = 2*np.random.random(size=(N, 2))-1

c = 0
for cor in xy_ran[:,]:
    c += circle(cor)

I=4*c/N
print('Integration of the unit radius circle:', I, ' Error:', abs(I-np.pi))

# 10D sphere
def sphere(r):
    if r<1:
        return 1
    else:
        return 0
    
D = 10    
xy_ran = 2*np.random.random(size=(N, D))-1
V = 2**10

c = 0
for cor in xy_ran[:,]:
    r = (cor**2).sum()
    c += sphere(r)
    
I_sp = V*c/N
print('Integration of the 10D sphere:', I_sp, ' Error:', abs(I_sp-np.pi**5/(120)))
```

5\. **Monte Carlo integration with importance sampling** 

Calculate the value of the integral:

$$
I=\int_0^1 \frac{x^{-1/2}}{e^x+1} dx
$$

using the importance sampling method with $w(x)=1/\sqrt{x}$. You should get a result about 0.84

```python
def f(x):
    return 1/((np.exp(x)+1)*np.sqrt(x))

def w(x):
    return 1/np.sqrt(x)

N = 100000
expect_res = 0.84

# Integration of w, Mean Montecarlo method
w_ran = w(np.random.rand(N))
w_integration = w_ran.sum()/N

# Importnce Sampling method
ran_samp = np.random.rand(N)
fw_ran = f(ran_samp)/w(ran_samp)
I = w_integration*fw_ran.sum()/N
print('Importance sampling method result:', I, ' Error respect to the expected result:', abs(I-expect_res))
```
