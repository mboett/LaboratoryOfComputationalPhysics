1\. **PCA on 3D dataset**

* Generate a dataset with 3 features each with N entries (N being ${\cal O}(1000)$). With $N(\mu,\sigma)$ the normali distribution with mean $\mu$ and $\sigma$  standard deviation, generate the 3 variables $x_{1,2,3}$ such that:
    * $x_1$ is distributed as $N(0,1)$
    * $x_2$ is distributed as $x_1+N(0,3)$
    * $x_3$ is given by $2x_1+x_2$
* Find the eigenvectors and eigenvalues of the covariance matrix of the dataset
* Find the eigenvectors and eigenvalues using SVD. Check that they are two procedure yields to same result
* What percent of the total variability is explained by the principal components? Given how the dataset was constructed, do these make sense? Reduce the dimensionality of the system so that at least 99% of the total variability is retained.
* Redefine the data in the basis yielded by the PCA procedure
* Plot the data points in the original and the new coordiantes as a set of scatter plots. Your final figure should have 2 rows of 3 plots each, where the columns show the (0,1), (0,2) and (1,2) proejctions.


```python

import numpy as np
from scipy import linalg as la
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

# Generate a dataset with 3 features each with N entries
N = 2000
sigma = 3

x1 = np.random.randn(N)
x2 = x1+sigma*np.random.randn(N)
x3 = 2*x1+x2

dtst = np.vstack((x1,x2,x3))

# Find the eigenvectors and eigenvalues of the covariance matrix of the dataset
covmat=np.cov(dtst)
print("Covariance matrix:\n", covmat)
e, V = la.eig(covmat)
print("\nEigenvalues of the covariance matrix:\n", e)
print("\nEigenvectors of the covariance matrix:\n", V)

# Find the eigenvectors and eigenvalues using SVD
U, S, Vt = la.svd(dtst)
SVD_eig = S**2/(N-1)
print("\nEigenvalues of the covariance matrix SVD:\n", SVD_eig)
print("\nEigenvectors of the covariance matrix SVD:\n", U)


# Reduce the dimensionality of the system so that at least 99% of the total variability is retained
variab=((SVD_eig[0]+SVD_eig[1])/SVD_eig.sum())
print("\nVariability:", variab)
# All the variability come from the first two eigenvectors, this is probably due to the fact that x3 is a linear combination of x1 and x2

# Redefine the data in the basis yielded by the PCA procedure
new_dtst = U.T.dot(dtst)

# Plot the data points in the original and the new coordiantes as a set of scatter plots
# Original basis
plt.figure(figsize=(18, 10))
ax1 = plt.subplot(2, 3, 1)
ax1.scatter(dtst[0,:], dtst[1,:], alpha=0.3)
ax1.set_title('Original basis: 0 and 1')
ax1.set_ylim(-20, 20)
ax1.set_xlim(-20, 20)

ax2 = plt.subplot(2,3,2)
ax2.scatter(dtst[0,:], dtst[2,:], alpha=0.3)
ax2.set_title('Original basis: 0 and 2')
ax2.set_ylim(-20, 20)
ax2.set_xlim(-20, 20)

ax3 = plt.subplot(2,3,3)
ax3.scatter(dtst[1,:], dtst[2,:], alpha=0.3)
ax3.set_title('Original basis: 1 and 2')
ax3.set_ylim(-20, 20)
ax3.set_xlim(-20, 20)

# New basis
ax4 = plt.subplot(2,3,4)
ax4.scatter(new_dtst[0,:], new_dtst[1,:], alpha=0.3)
ax4.set_title('New basis: 0 and 1')
ax4.set_ylim(-20, 20)
ax4.set_xlim(-20, 20)

ax5 = plt.subplot(2,3,5)
ax5.scatter(new_dtst[0,:], new_dtst[2,:], alpha=0.3)
ax5.set_title('New basis: 0 and 2')
ax5.set_ylim(-20, 20)
ax5.set_xlim(-20, 20)

ax6 = plt.subplot(2,3,6)
ax6.scatter(new_dtst[1,:], new_dtst[2,:], alpha=0.3)
ax6.set_title('New basis: 1 and 2')
ax6.set_ylim(-20, 20)
ax6.set_xlim(-20, 20)

plt.show()
```

2\. **PCA on a nD dataset**

Start from the dataset you have genereted in the previous exercise and add uncorrelated random noise. Such noise should be represented by other 10 uncorrelated variables normal distributed, with standar deviation much smaller (say, a factor 50) than those used to generate the $x_1$ and $x_2$.

Repeat the PCA procedure and compare the results with what you obtained before

```python
dtst_noisy = dtst
for i in range(10):
    xNoise = (np.random.uniform(low=1.0, high=sigma)/50*np.random.randn(N))
    dtst_noisy = np.vstack((dtst_noisy,xNoise))
    
U, S, Vt = la.svd(dtst_noisy)
SVD_eig = S**2/(N-1)
print("\nEigenvalues:\n", SVD_eig)

variab=((SVD_eig[0]+SVD_eig[1])/SVD_eig.sum())
print("\nVariability:", variab)

new_dtst_noisy = U.T.dot(dtst_noisy)

c = 1
plt.figure(figsize=(18, 100))
for i in range(10):
    for j in range(i+1,11):
        ax = plt.subplot(19, 3, c)
        ax.scatter(new_dtst_noisy[i,:],new_dtst_noisy[j,:], alpha=0.3)
        ax.axis([-20, 20, -20, 20])
        ax.set_title("Noisy dataset new basis: "+str(i)+" and "+str(j))
        c = c + 1
        
plt.show()
```
